# Análisis de una solución de la ecuación del calor

La solución de la ecuación
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2}$$
sujeta a las condiciones de frontera:
$$u(0,t)=u(1,t)=0, \;\;\; t>0$$
Y la condición inicial:
$$u(x,0)=x,\;\;\;0<x<1$$
Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,t)=\frac{2}{\pi}\sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n}  \sin(n\pi x)e^{-(n\pi)^2 t}.$$

Este problema modela la transferencia de calor en una barra ideal delgada, con $k=1$.

# Explicación
## Pasos para Resolver la Ecuación de Calor por el Método de Separación de Variables

1. **Planteamiento de la Ecuación de Calor:**
   $$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2}$$
   Sujeta a las condiciones de frontera y condición inicial:
   - $$u(0,t)=u(1,t)=0, \; t>0$$
   - $$u(x,0)=x, \; 0<x<1$$

2. **Suposición de Separación de Variables:**
   Se asume que la solución es de la forma:
   $$u(x, t) = X(x)T(t)$$

3. **Sustitución en la Ecuación de Calor:**
   $$X(x)T'(t) = X''(x)T(t)$$

4. **Separación de Variables:**
   Reorganizamos para obtener:
   $$\frac{T'(t)}{T(t)} = \frac{X''(x)}{X(x)} = -\lambda$$

5. **Resolución de las Ecuaciones Diferenciales:**
   - Para la parte espacial:
     $$X''(x) + \lambda X(x) = 0$$
   - Para la parte temporal:
     $$T'(t) + \lambda T(t) = 0$$

6. **Condiciones de Frontera y Obtención de Valores Propios:**
   La condición de frontera $$u(0,t) = u(1,t) = 0$$ lleva a:
   $$X(0) = X(1) = 0$$
   Resolviendo el problema de Sturm-Liouville, obtenemos:
   $$X_n(x) = \sin(n\pi x), \; \lambda_n = (n\pi)^2$$

7. **Solución General:**
   La solución se expresa como:
   $$u(x,t) = \sum_{n=1}^{\infty} B_n \sin(n\pi x) e^{-(n\pi)^2 t}$$

8. **Determinación de los Coeficientes:**
   Usamos la condición inicial:
   $$u(x,0) = x$$
   Aplicamos el método de coeficientes de Fourier:
   $$B_n = \frac{2}{\pi} \frac{(-1)^{n+1}}{n}$$

9. **Solución Final:**
   $$u(x,t) = \frac{2}{\pi}\sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n} \sin(n\pi x)e^{-(n\pi)^2 t}$$



```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y
```

```python
from pylab import *
```

```python
x = linspace(0,1,1000)
f = recta(x)
plot(x,f,label='$u(x,0)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-0.01,1.01]),ylim([-0.01,1.01]),grid(),legend()
show()
```

Usamos una serie de Fourier para aproximar la condición inicial $u(x,0)=f(x)=x$.

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (-1)**(n+1)*sin(n*pi*x)/(n)
    return (2.0/pi)*f
```

```python
f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.1]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Podemos visualizar la aproximación para diferentes cantidades de armónicos.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Ahora, agregamos la parte temporal para visualizar la solución completa. Primero, para el momento inicial, $t=0$.

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += (-1)**(n+1)*sin(n*pi*x)/(n)*exp(-n**2*pi**2*t[i])
    return (2.0/pi)*f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
```

```python
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Luego, para diferentes tiempos, lo cual nos permite ver cómo la solución evoluciona cuando $t \to \infty$.

```python
plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evolución de la solución")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Y podemos ver esto en una gráfica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
t = linspace(0,1.0,100)

X,T = np.meshgrid(x,t)

U = u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$');
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="plasma"); #Otros colores:"viridis", "inferno", "plasma"
```
