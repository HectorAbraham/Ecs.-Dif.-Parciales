```python
from sympy import fourier_series, pi, plot, Piecewise, symbols, exp
from sympy.abc import x
```

```python
f = exp(2*x)
```

```python
plot(f, (x,0,1),line_color ='blue')
```

```python
s = fourier_series(f, (x, 0, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 10)
p = plot(f, s1, s2, s3, (x, 0, 1), show=False, legend=True)

s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=10'

p.show()
```

```python
g = Piecewise(
    (1, (x >= 0) & (x <= pi/2)),  # f(x) = 1 para 0 <= x <= pi/2
    (2, (x > pi/2) & (x <= pi))   # f(x) = 2 para pi/2 < x <= pi
)
```

```python
plot(g, (x,0,pi),line_color ='red')
```

```python
s = fourier_series(g, (x, 0, pi))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 10)
p = plot(f, s1, s2, s3, (x, 0, pi), show=False, legend=True)

s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=10'

p.show()
```
